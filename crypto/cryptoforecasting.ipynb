```python
volume_cumsum = 0
price_volume_sum =0
first = True
vwap_list=[]
for i in range(len(binance_bitcoin)):
    # 가중합 대표가격을 구하기 위해서 고가+ 저가 + 종가 나누기 3 을한다
    represent_price = (binance_bitcoin.iloc[i]['High'].max()+ binance_bitcoin.iloc[i]['Low'].min() + binance_bitcoin.iloc[i]['Close'])/3

    volume_cumsum += binance_bitcoin.iloc[i]['Volume'].sum()
    # vwap = (가격* 거래량)의 합 / 거래량의 합

    # 해당 기간 ( 일 ) 동안의 거래량을 곱한다 대표가격 * 거래량
    price_volume_sum  += represent_price *  binance_bitcoin.iloc[i]['Volume'].sum()
    # 이때까지의 총 거래량으로 나눈다 총 거래량으 어디부터 재느냐에 따라 다르다 # (누적 가격 * 거래량) / (누적 거래량)
    vwap = price_volume_sum / volume_cumsum
    vwap_list.append(vwap)
```

```python
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            tmp_mdl  = SARIMAX(target.values[:1000], exog=None , order= param,
                                                 seasonal_order = param_seasonal,
                                                 enforce_stationarity=True,
                                                 enforce_invertibility=True)
            res =tmp_mdl.fit()
            print("SARIMAX{}x{}12 - AIC : {}".format(param, param_seasonal,res.aic))
            if res.aic < best_aic:
                best_aic = res.aic
                best_pdq = param
                best_seasonal_pdq = param_seasonal
                best_mdl = tmp_mdl
        except:
            #print("Unexpected error:",sys.sys.exc_info[0])
            continue
print("Best SARIMAX{}x{}12 model - AIC:{}".format(best_pdq, best_seasonal_pdq, best_aic))
```
